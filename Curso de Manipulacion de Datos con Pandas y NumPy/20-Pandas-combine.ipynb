```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
})
```

```python
pd.concat([df1, df2], ignore_index=True) #default axis 0
# by default index is inherited
```

```python
pd.concat([df1, df2], axis=1)
```

# Merge

```python
izq = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'a': ['a0', 'a1', 'a2', 'a3'],
    'b': ['b0', 'b1', 'b2', 'b3'],
})

der = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'c': ['c0', 'c1', 'c2', 'c3'],
    'd': ['d0', 'd1', 'd2', 'd3'],
})
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'a': ['a0', 'a1', 'a2', 'a3'],
    'b': ['b0', 'b1', 'b2', 'b3'],
})

der = pd.DataFrame({
    'key_2': ['k0', 'k1', 'k2', np.nan],
    'c': ['c0', 'c1', 'c2', 'c3'],
    'd': ['d0', 'd1', 'd2', 'd3'],
})
```

```python
izq.merge(der, left_on='key', right_on='key_2') # inner join
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='left') # left
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='right') # right join
```
